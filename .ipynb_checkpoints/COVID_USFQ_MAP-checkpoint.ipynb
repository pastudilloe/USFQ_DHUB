```python
import pandas as pd

from keplergl import KeplerGl
import geopandas as gpd
```

```python
path = ("/Users/pastudilloe/Dropbox/USFQ_COVID/Data/")
```

```python
raw_data = pd.read_csv(path+ "CoV19.csv")
city_coord = pd.read_csv(path+ "EC_cities_coordinates.csv")
```

```python
raw_data
```

```python
data = raw_data.groupby(['Time_N', 'city_code', 'Linaje', 'Clado']).agg({'GISAID ID':'count'}).reset_index()
data
```

```python
data = pd.merge(data, city_coord, how="left", left_on="city_code", right_on='city_code')
data['datetime'] = data['Time_N'].astype(str) + ' 0:00'
data
```

```python
gdf = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data.point_x, data.point_y))
```

```python
gdf
```

```python
#Create a basemap 
map = KeplerGl(height=600, width=800)
```

```python
map.add_data(data=gdf, name='canton_name')

#show the map
map
```
