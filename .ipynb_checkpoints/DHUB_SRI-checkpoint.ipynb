```python
canton_SRI= pd.read_csv(path+'Files Open Data/sri_recaudacion_diccionario.csv', encoding='latin-1')
canton_SRI.columns = ['PROVINCIA', 'CANTON', 'COD_Canton']

Pobl_Cantones = pd.read_csv(path+'/GENERALES/EC_canton_poblacion_latlon.csv')
Pobl_Cantones.columns = ['Provincia', 'Canton', 'COD_Canton', 'point_x', 'point_y', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', 'Pop_2020', '2021', '2022', '2023', '2024', '2025', '2026', '2027', '2028', '2029', '2030']
Pobl_Cantones = Pobl_Cantones[['Provincia', 'Canton', 'COD_Canton','point_x', 'point_y', 'Pop_2020']]
```

### DATA PROCESSING

```python
rec_2021_ = pd.merge(rec_2021, canton_SRI, how='left', on=['PROVINCIA', 'CANTON'])
```

```python
rec_2021_1 = rec_2021_.groupby(['ANIO', 'MES', 'GRUPO_IMPUESTO' ,'SECTOR_N1', 'TIPO_CONTRIBUYENTE', 'PROVINCIA', 'CANTON', "COD_Canton"]).agg({'VALOR_RECAUDADO':'sum'}).reset_index()
rec_2021_1.sort_values(by='VALOR_RECAUDADO')
```

```python
rec_2021_2 = rec_2021_1.groupby(['ANIO', 'PROVINCIA', 'CANTON', 'COD_Canton']).agg({'VALOR_RECAUDADO':'sum'}).reset_index()
rec_2021_2 = rec_2021_2.sort_values(by='VALOR_RECAUDADO', ascending = True)

rec_2021_3 = pd.merge(rec_2021_2, Pobl_Cantones, how='left', on='COD_Canton')

rec_2021_3['REC_PerCapita'] = rec_2021_3['VALOR_RECAUDADO'] / rec_2021_3['Pop_2020']
rec_2021_3['PROV_CANT'] = rec_2021_3['PROVINCIA'] + "-" + rec_2021_3['CANTON']

```

```python
rec_2021_3_top = rec_2021_3.sort_values(by='VALOR_RECAUDADO', ascending = False).head(20)
sns.barplot(data=rec_2021_3_top, x="VALOR_RECAUDADO", y="PROV_CANT")
```

```python
rec_2021_3_top = rec_2021_3.sort_values(by='REC_PerCapita', ascending = False).head(20)
sns.barplot(data=rec_2021_3_top, x="REC_PerCapita", y="PROV_CANT")
```

```python
rec_2021_3_bottom = rec_2021_3.sort_values(by='VALOR_RECAUDADO', ascending = True).head(20)
sns.barplot(data=rec_2021_3_bottom, x="VALOR_RECAUDADO", y="PROV_CANT")
```

```python
rec_2021_3_bottom = rec_2021_3.sort_values(by='REC_PerCapita', ascending = True).head(20)
sns.barplot(data=rec_2021_3_bottom, x="VALOR_RECAUDADO", y="PROV_CANT")
```
